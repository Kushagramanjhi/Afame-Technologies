```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("/content/ECOMM DATA.xlsx")
```

```python
df.info()
```

```python
df = df.sort_values("Order Date")
```

```python
salesday = df.groupby("Order Date")["Sales"].sum()
```

```python
salesday.plot(figsize=(60, 18))
```

```python
salesmonth = salesday.resample("M").sum()
```

```python
salesmonth
```

```python
salesmonth.plot(kind='line', title='Monthly Sales', figsize=(10, 6))
plt.ylabel('Sales')
plt.xlabel('Month')
plt.show()
```

```python
salesyear = salesday.resample("Y").sum()
```

```python
salesyear.plot(kind='bar', title='Yearly Sales', figsize=(10, 6))
plt.ylabel('Sales')
plt.xlabel('Year')
```

```python
Profits = df.groupby("Category")["Profit"].sum()
```

```python
Profits.plot(kind='bar', title='Category Profits', figsize=(10, 6))
plt.ylabel('Profits')
plt.xlabel('Category')
```

```python
Profits2 = df.groupby("Sub-Category")["Profit"].sum()
```

```python
Profits2.plot(kind='bar', title='Sub-Category Profits', figsize=(10, 6))
plt.ylabel('Profits')
plt.xlabel('Sub-Category')
```

```python
profits3 = df.groupby("Sub-Category")["Quantity"].sum()
```

```python
profits3.plot(kind='bar', title='Sub-Category Quantity', figsize=(10, 6))
plt.ylabel('Quantity')
plt.xlabel('Sub-Category')
```
